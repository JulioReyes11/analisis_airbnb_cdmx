## Paso 3: Desarrollo.

```python
# Cargar datos limpios y ver la estructura de datos.  
df = pd.read_csv('../datos/datos_filtrados.csv')
df.head()
```

```python
df.shape
```

## 3.1 Total de alojamientos por neighborhood (Alcaldía) 

```python
#Total de alojamientos por alcaldía
alojamientos_por_alcaldia = df['neighbourhood'].value_counts()
alojamientos_por_alcaldia
```

```python
# Gráfica de alojamientos por alcaldía
ax = sns.barplot(x= alojamientos_por_alcaldia, y= alojamientos_por_alcaldia.index, orient='h')
ax.set_title('Total de alojamientos por alcaldías')
ax.set(xlabel='Total', ylabel='Alcaldía');
```

## 3.2 Distribución espacial de alojamientos (Mapa de calor)

```python
# Obtener coordenadas para centrar el mapa
media_latitude = df['latitude'].mean()
media_longitude = df['longitude'].mean()
print(media_latitude)
print(media_longitude)
```

```python
# Mapa interactivo
map_folium = folium.Map([19.397890,-99.167112],zoom_start=9, width='100%', height='100%')
HeatMap(df[['latitude','longitude']].dropna(),radius=10).add_to(map_folium)
display(map_folium)
```

## 3.3 Tipo de alojamiento mas ofertado

```python
# Alojamiento más ofertado
total_por_habitacion = df['room_type'].value_counts()
total_por_habitacion
```

```python
#Gráfica de alojamiento más ofertado
ax = sns.barplot(x = total_por_habitacion.index, y = total_por_habitacion)
ax.set_title('Total por tipo de habitación')
ax.set(ylabel='Total', xlabel= 'Habitación');
```

## 3.4 Distribución de precios por alojamientos

```python
sns.distplot(x=df['price'], norm_hist=False, bins=50)
```

```python
print(f'Curtosis: {kurtosis(df.price)}')
print(f'Asimetría: {skew(df.price)}')
```

```python
# Resumen de estadísticas precios 
df['price'].describe()
```

## 3.5 Precios promedio por alcaldía

### 3.5.1 Precio promedio de Entire home/apt por alcaldía
Entire home/apt

```python
# Filtro y agrupamiento por precio promedio de alojamiento completo
filtro_alojamiento_completo = df['room_type'] == 'Entire home/apt'
df_filtro_alojamiento_completo = df[filtro_alojamiento_completo]
promedio_alojamiento_completo = df_filtro_alojamiento_completo.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
promedio_alojamiento_completo
```

```python
#Gráfica de alojamiento completo Entire home/apt precio promedio
ax = sns.barplot(x= promedio_alojamiento_completo, y= promedio_alojamiento_completo.index, orient='h')
ax.set_title('Precio promedio de Entire home/apt')
ax.set(xlabel='Precio Promedio', ylabel='Alcaldía');
```

### 3.5.2 Precio promedio de Private room por alcaldía
Private room

```python
# Filtro y agrupamiento por precio promedio de Private room
filtro_alojamiento_privado = df['room_type'] == 'Private room'
df_filtro_alojamiento_privado = df[filtro_alojamiento_privado]
promedio_alojamiento_privado = df_filtro_alojamiento_privado.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
promedio_alojamiento_privado
```

```python
#Gráfica de alojamiento completo Private room precio promedio
ax = sns.barplot(x= promedio_alojamiento_privado, y= promedio_alojamiento_privado.index, orient='h')
ax.set_title('Precio promedio de Private room')
ax.set(xlabel='Precio Promedio', ylabel='Alcaldía');
```

### 3.5.3 Precio promedio de Shared room por alcaldía
Shared room (habitación compartida)

```python
# Filtro y agrupamiento por precio promedio de alojamiento Shared room
filtro_alojamiento_compartido = df['room_type'] == 'Shared room'
df_filtro_alojamiento_compartido = df[filtro_alojamiento_compartido]
promedio_alojamiento_compartido = df_filtro_alojamiento_compartido.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
promedio_alojamiento_compartido
```

```python
#Gráfica de alojamiento completo Shared roo precio promedio
ax = sns.barplot(x= promedio_alojamiento_compartido, y= promedio_alojamiento_compartido.index, orient='h')
ax.set_title('Precio promedio de Shared room')
ax.set(xlabel='Precio Promedio', ylabel='Alcaldía');
```

### 3.5.4 Precio promedio de Hotel room por alcaldía
Hotel room

```python
# Filtro y agrupamiento por precio promedio de alojamiento cuarto hotel
filtro_alojamiento_hotel = df['room_type'] == 'Hotel room'
df_filtro_alojamiento_hotel = df[filtro_alojamiento_hotel]
promedio_alojamiento_hotel = df_filtro_alojamiento_hotel.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
promedio_alojamiento_hotel
```

```python
#Gráfica de alojamiento completo Entire Hotel room precio promedio
ax = sns.barplot(x= promedio_alojamiento_hotel, y= promedio_alojamiento_hotel.index, orient='h')
ax.set_title('Precio promedio de Hotel room')
ax.set(xlabel='Precio Promedio', ylabel='Alcaldía');
```

## Paso 4: Conclusiones.

- De los datos analizados se contabilizaron 18, 475 lugares ofertados. Las alcaldías con mayor presencia de alojamientos son: en primer lugar se encuentra Cuauhtémoc con 6,941, en segundo lugar esta Benito Juárez con 2,987 y en tercer lugar se encuentra la alcaldía Miguel Hidalgo con 2,831. 


- Se puede notar que los alojamientos se concentran en lugares con mayor desarrollo inmobiliario y disponibilidad de servicios. Habría que analizar la correlación con la distancias de el aeropuerto, estaciones de metro, terminales de autobuses y lugares turísticos.


- El análisis nos muestra que el apartamento y casa completa (Entire home/apt) es el alojamiento más ofertado con 9,412 anuncios y la habitación privada (Private room) con 8,507 anuncios y en tercer lugar la habitación  compartida (shared room) con 358 anuncios y por último la habitación de hotel (hotel room) con 198 anuncios.


- La distribución de los precios de los alojamientos de Airbnb en la Ciudad de México tiene una distribución con asimetría positiva de un valor 1.05, con una cola ligeramente larga que tiene como curtosis 0.48. El precio promedio es `$751.61`, los precios tienen un valor mínimo de `$100` y un valor máximo de `$2,884`.

    Los valores atípicos de la variable precios fueron eliminados, había valores de 0 y valores máximos de `$350,000` que generaban ruido a nuestro conjunto de datos.

- Los precios promedio por tipo de alojamiento resultaron diferentes para las alcaldías:

    Para el caso de habitación y departamento completo la alcaldía Miguel Hidalgo es la más alta con un precio promedio de `$1,162` y después le sigue Cuajimalpa de Morelos y Cuauhtémoc con `$1,095` y `$1,037` respectivamente.

    Para las habitaciones privadas encontramos que Milpa Alta es la más alta con un precio promedio de `$829`, Cuajimalpa de Morelos se mantiene en el segundo lugar con un valor de `$599` y en tercer lugar se encuentra la alcaldía Miguel Hidalgo con un valor de `$583`.

    El precio promedio por habitación compartida más alto se encuentra en Cuajimalpa de Morelos con un valor de `$802`, después encontramos valores cerca de los `$500`.

    Y por ultimo la habitación de hotel se reduce a 9 alcaldías con anuncios disponibles, con Cuajimalpa de Morelos con el precio promedio más alto con un valor de `$1,721`.
